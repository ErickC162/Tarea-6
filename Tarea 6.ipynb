### Utilizando Series de Taylor

#### 1. $\frac{1}{25 * x^{2} + 1},\ x_0=0$

Orden de la mejor aproximación:

La serie de Taylor converge rápidamente para valores de x cercanos a 0. Un orden de 2 o 3 ya da una buena aproximación en este intervalo.

Fórmulas de los polinomios:

n = 0: 1

n = 1: 1 - 25x²

n = 2: 1 - 25x² + 625x⁴

... (sigue el patrón)

```python
import numpy as np
import matplotlib.pyplot as plt

def taylor_approx_1(x, n):
    """Aproximación de Taylor de orden n para 1/(25x^2 + 1) en x0 = 0"""
    approx = 0
    for i in range(n + 1):
        approx += (-25)**i * x**(2*i)
    return approx

# Gráfica
x = np.linspace(-0.4, 0.4, 400)  # Rango de x centrado en 0
plt.figure(figsize=(10, 6))

# Función original
plt.plot(x, 1 / (25 * x**2 + 1), label="1/(25x^2 + 1)")

# Aproximaciones de Taylor (diferentes órdenes)
for n in [0, 1, 2, 3, 5]:
    plt.plot(x, taylor_approx_1(x, n), label=f"Taylor (n={n})")

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Aproximaciones de Taylor para 1/(25x^2 + 1)")
plt.legend()
plt.grid(True)
plt.show()
```

#### 2. $arctan(x),\ x_{0}=1$

Orden de la mejor aproximación:

La convergencia es más lenta que en el caso anterior. Para una buena aproximación en el intervalo \[0, 2], necesitarás un orden mayor, como 7 o 9.

Fórmulas de los polinomios:

n = 1: π/4 + (x - 1)/2

n = 3: π/4 + (x - 1)/2 - (x - 1)³/6

n = 5: π/4 + (x - 1)/2 - (x - 1)³/6 + (x - 1)⁵/10

... (sigue el patrón)

```python
def taylor_approx_arctan(x, n):
    """Aproximación de Taylor de orden n para arctan(x) en x0 = 1"""
    approx = np.pi/4
    for i in range(1, n + 1):
        approx += (-1)**(i - 1) * (x - 1)**i / (2*i)
    return approx

# Gráfica
x = np.linspace(0, 2, 400)  # Rango de x centrado en 1
plt.figure(figsize=(10, 6))

# Función original
plt.plot(x, np.arctan(x), label="arctan(x)")

# Aproximaciones de Taylor (diferentes órdenes)
for n in [1, 3, 5, 7, 9]:
    plt.plot(x, taylor_approx_arctan(x, n), label=f"Taylor (n={n})")

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Aproximaciones de Taylor para arctan(x)")
plt.legend()
plt.grid(True)
plt.show()
```

### Usando polinomios de Lagrange

#### 1. $\frac{1}{25 * x^{2} + 1},\ x_0=0$
Orden de la mejor aproximación:

Para determinar el orden de la mejor aproximación, necesitamos conocer más información sobre los puntos de interpolación. El polinomio de Lagrange de grado n interpola exactamente una función en n+1 puntos. Si solo tenemos x₀ = 0, el mejor polinomio de Lagrange que podemos construir es de grado 0, es decir, una constante.

Fórmula del polinomio:

El polinomio de Lagrange de grado 0 (constante) que interpola f(x) en x₀ = 0 es simplemente:

$P_0(x) = f(0) = \frac{1}{25 * 0^{2} + 1} = 1$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 1 / (25 * x**2 + 1)

x = np.linspace(-1, 1, 100)  # Rango para graficar
y = f(x)
y_approx = np.ones_like(x)  # Aproximación constante

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = 1/(25x² + 1)')
plt.plot(x, y_approx, label='P₀(x) = 1', linestyle='dashed')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Aproximación de Lagrange (grado 0)')
plt.legend()
plt.grid(True)
plt.show()
```

#### 2. $arctan(x),\ x_{0}=1$
Orden de la mejor aproximación:

Similar al caso anterior, con un solo punto (x₀ = 1), la mejor aproximación será una constante.

Fórmula del polinomio:

$P_0(x) = f(1) = arctan(1) = \pi/4$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.arctan(x)

x = np.linspace(0, 2, 100)  # Rango para graficar
y = f(x)
y_approx = np.full_like(x, np.pi/4)  # Aproximación constante

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = arctan(x)')
plt.plot(x, y_approx, label='P₀(x) = π/4', linestyle='dashed')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Aproximación de Lagrange (grado 0)')
plt.legend()
plt.grid(True)
plt.show()
```
